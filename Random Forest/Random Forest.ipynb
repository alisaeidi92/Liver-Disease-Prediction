```python
def set_outlier(col,max_value, set_value, min_age = 0):
    
    #col: The column in data
    #max_value: the threshold of outliers
    #set_value: the highest possible value for instance
    
    for i,instance in enumerate(liver_df[col]):
        
        if liver_df['Age'].iloc[i] >= min_age:
            
            if instance >= max_value:
                liver_df[col].iloc[i] = set_value
            
```

```python
#An elevation greater than 20 mg/dL suggests severe liver disease. 
#In patients with hepatitis-induced acute liver failure, 
#a serum total bilirubin level > 17.5 mg/dL (300 mmol/L) is a criterion for predicting death and the need for liver transplantation.

set_outlier('Total_Bilirubin', max_value= 20, set_value= 20)


#Typically, bilirubin levels fall somewhere between 0.3 and 1.2 milligrams per deciliter (mg/dL). 
#Anything above 1.2 mg/dL is usually considered high. 
#The condition of having high bilirubin levels is called hyperbilirubinemia

set_outlier('Direct_Bilirubin', max_value= 10, set_value= 10)


#The ALP level in healthy adults should be 20–140 units per liter (U/L). 
#Children tend to have significantly higher levels of ALP than adults because their bones are still growing. 
#A person recovering from a bone injury may also have a raised ALP level in the 3 months after the injury while their bone heals.


set_outlier('Alkaline_Phosphotase', max_value= 1000, set_value= 1000, min_age= 18)


#Typically the range for normal AST is reported between 10 to 40 units per liter and ALT between 7 to 56 units per liter. 
#Mild elevations are generally considered to be 2-3 times higher than the normal range. 
#In some conditions, these enzymes can be severely elevated, in the 1000s range

set_outlier('Alamine_Aminotransferase', max_value= 1000, set_value= 1000)


#Typically the range for normal AST is reported between 10 to 40 units per liter and ALT between 7 to 56 units per liter. 
#Mild elevations are generally considered to be 2-3 times higher than the normal range. 
#In some conditions, these enzymes can be severely elevated, in the 1000s range.

set_outlier('Aspartate_Aminotransferase', max_value= 1000, set_value= 1000)


#Total protein range. The normal range for total protein is between 6 and 8.3 grams per deciliter (g/dL). 
#This range may vary slightly among laboratories.

#Total Protein Plot seems OK



#Results are given in grams per deciliter (g/dL). A normal albumin range is 3.4 to 5.4 g/dL. 
#If you have a lower albumin level, you may have malnutrition. 
#It can also mean that you have liver disease or an inflammatory disease.


#Albumin plot seems OK


#A normal range of albumin is 39 to 51 grams per liter (g/L) of blood. 
#The normal range for globulins varies by specific type. A normal range for total globulins is 23 to 35 g/L. 
#If your protein level is low, you may have a liver or kidney problem.

# A/G Ratio seems OK



```

# Age groups

```python
liver_df['child/teen'] = 0
liver_df['young'] = 0
liver_df['mid-age'] = 0
liver_df['old'] = 0
```

```python
for i,instance in enumerate(liver_df['Age']):
    
    if instance < 18:
        liver_df['child/teen'].iloc[i] = 1
    elif 18 <= instance <= 40:
        liver_df['young'].iloc[i] = 1
    elif 40 < instance <= 60:
        liver_df['mid-age'].iloc[i] = 1
    elif 60 < instance:
        liver_df['old'].iloc[i] = 1
```

```python
liver_df.drop(['child/teen'], axis = 1, inplace=True)
```

```python
liver_df.drop(['Age'], axis = 1, inplace = True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_mm, y, test_size=0.15, random_state=42)
```

# Using GridSearchCV to find the best parameters for Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
estimators = range(1,100,8)
criterion = ['gini', 'entropy']
min_samples_split = range(1,8)
max_features = ['auto', 'sqrt', 'log2']



# Create hyperparameter options
hyperparameters = dict(n_estimators=estimators,criterion = criterion,min_samples_split = min_samples_split, max_features=max_features)
```

```python
rf = RandomForestClassifier(verbose=1, random_state=42)
```

```python
clf = GridSearchCV(rf, hyperparameters,scoring='f1', cv=5, verbose=1)
```

```python
best_model = clf.fit(X_train, y_train)
```

```python
print('Best n_estimators:', best_model.best_estimator_.get_params()['n_estimators'])
print('Best criterion:', best_model.best_estimator_.get_params()['criterion'])
print('Best max_features:', best_model.best_estimator_.get_params()['max_features'])
print('Best min_samples_split:', best_model.best_estimator_.get_params()['min_samples_split'])
```

# Training with the best parameters

```python
tuned_rf = RandomForestClassifier(n_estimators=9,
                                  criterion='gini',
                                  max_features='auto'
                                  ,min_samples_split=7)
```

```python
tuned_rf.fit(X_train,y_train)
```

```python
y_preds = tuned_rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_preds))
```
