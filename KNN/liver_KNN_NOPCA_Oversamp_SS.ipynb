**Using SMOTE to remove the imblance in data**

```

from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
os = SMOTE(random_state=0)
X_train, X_test, y_train, y_test = train_test_split(liver_df, labels, test_size=0.3, random_state=0)
columns = X_train.columns
os_data_X,os_data_y=os.fit_sample(X_train, y_train)
os_data_X = pd.DataFrame(data=os_data_X,columns=columns )
os_data_y= pd.DataFrame(data=os_data_y,columns=['Liver_Disease'])
# we can Check the numbers of our data
print("length of oversampled data is ",len(os_data_X))
print("Number of no liver disease in oversampled data",len(os_data_y[os_data_y['Liver_Disease']==0]))
print("Number of  liver disease in oversampled data",len(os_data_y[os_data_y['Liver_Disease']==1]))
print("Proportion of no liver disease in oversampled data is ",len(os_data_y[os_data_y['Liver_Disease']==0])/len(os_data_X))
print("Proportion of  liver disease data in oversampled data is ",len(os_data_y[os_data_y['Liver_Disease']==1])/len(os_data_X))
```

```
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import mean_squared_error 
from math import sqrt
from sklearn.metrics import roc_auc_score
import itertools

accuracy_score1 = []
def KNN_value(metric):
  rmse_val1 = []
  for q in range(25):
    K_value = q+1
    tuned_KNN1 = KNeighborsClassifier(n_neighbors = K_value, weights = 'uniform' , metric = metric )
    tuned_KNN1.fit(train, train_labels)
    knn_pred1 = tuned_KNN1.predict(test)
    error1 = sqrt(mean_squared_error(test_labels,knn_pred1)) #calculate rmse
    rmse_val1.append(error1) #store rmse values
  plt.figure(figsize =(10, 6)) 
  plt.plot(range(1, 26), rmse_val1, color ='blue', 
                linestyle ='dashed', marker ='o', 
         markerfacecolor ='red', markersize = 10) 
  
  plt.title('Error Rate vs. K Value' ) 
  plt.xlabel('K') 
  plt.ylabel('Error Rate') 
```

**Function to Calculate Accuracy , ROC curve , Confusion Matrix for different distance and K values**

```
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import mean_squared_error 
from math import sqrt
from sklearn.metrics import roc_auc_score
import itertools

accuracy_score1 = []
def KNN_model(metric,q):

  K_value =q
# Accuracy and F1-score using KNN model
  tuned_KNN1 = KNeighborsClassifier(n_neighbors = q, weights = 'uniform' ,metric = metric )
  tuned_KNN1.fit(train, train_labels)
  knn_pred1 = tuned_KNN1.predict(test)
  #print('\n\tK Nearest Neighbors (NN = 4)\n')
  accuracy_score = metrics.accuracy_score(test_labels, knn_pred1)*100
  accuracy_score1.append(accuracy_score)
  print('Accuracy Score: ',accuracy_score, '%for K-Value:',K_value,sep = '')
  cm1 = metrics.confusion_matrix(test_labels, knn_pred1)
  print('Confusion Matrix: ',metrics.confusion_matrix(test_labels, knn_pred1), sep = '\n')
  print('\n')
  print(classification_report(test_labels, knn_pred1))
  #error1 = sqrt(mean_squared_error(test_labels,knn_pred1)) #calculate rmse
  #rmse_val1.append(error1) #store rmse values
  #print('RMSE value for k= ' , K_value , 'is:', error1)
  print('\n')
  # Calculate roc auc
  #roc_value1 = roc_auc_score(y_test, knn_pred)
  from sklearn.metrics import roc_auc_score
  from sklearn.metrics import roc_curve
  roc_value1 = roc_auc_score(test_labels, knn_pred1)
  fpr, tpr, thresholds = roc_curve(test_labels, knn_pred1)
  plt.figure()
  plt.plot(fpr, tpr, label='KNN (area = %0.2f)' % roc_value1)
  plt.plot([0, 1], [0, 1],'r--')
  plt.xlim([0.0, 1.0])
  plt.ylim([0.0, 1.05])
  plt.xlabel('False Positive Rate')
  plt.ylabel('True Positive Rate')
  plt.title('Receiver operating characteristic')
  plt.legend(loc="lower right")
  plt.savefig('Log_ROC')
  plt.show()
  print(classification_report(test_labels, knn_pred1))

  # Plot the confusion matrix
  cmap=plt.cm.Oranges
  classes = ['0', '1']
  plt.figure(figsize = (6, 6))
  plt.imshow(cm1, interpolation='nearest', cmap=cmap)
  plt.title('Liver Disease CM', size = 24)
  plt.colorbar(aspect=4)
  tick_marks = np.arange(len(classes))
  plt.xticks(tick_marks, classes, rotation=45, size = 14)
  plt.yticks(tick_marks, classes, size = 14)

  #fmt = '.2f' if normalize else 'd'
  thresh = cm1.max() / 2.
    
 # Labeling the plot
  for i, j in itertools.product(range(cm1.shape[0]), range(cm1.shape[1])):
      plt.text(j, i, format(cm1[i, j]), fontsize = 20,
                horizontalalignment="center",
                color="white" if cm1[i, j] > thresh else "black")
        
  plt.grid(None)
  plt.tight_layout()
  plt.ylabel('Test label', size = 18)
  plt.xlabel('Predicted label', size = 18)

```
